# Panorama do COVID-19 no Brasil

COVID-19 é uma doença infecciosa causada por recém-descoberto coronavírus.

Transmitido principalmente por meio de gotículas provenientes de tosses ou espirros de pessoas infectadas, a gravidade dos sintomas varia muito de pessoa para pessoa.

Fato é, não se sabe muita coisa a respeito do COVID-19. Estudos estão sendo realizados no mundo todo, porém os resultados ainda não são conclusivos e definitivos.

<center><img alt="Colaboratory logo" width="50%" src="https://static.poder360.com.br/2022/12/coronavirus-imagem-1-848x477.jpg"></center>

Até o presente momento, observa-se que cerca de 80% dos casos confirmados são assintomáticos e rápidos. A maioria das pessoas que se encaixam nesse grupo, se recupera sem nenhuma sequela.

No entanto, 15% das pessoas terão infecções graves e precisarão de oxigênio. O restante das pessoas, que representam 5%, serão classificadas como infecções muito graves e precisarão de ventilação assistida, por meio de respiradores mecânicos em ambiente hospitalar.

Com o objetivo de elevar a consciência situacional a respeito do COVID-19 no Brasil, irei realizar uma análise sobre os dados públicos da doença.

### Dicionário de Variáveis
* codigo_iso: ISO 3166-1 alpha-3 – códigos de países com três letras. Observe que as regiões definidas pelo OWID (por exemplo, continentes como “Europa”) contêm o prefixo “OWID_”.
* continente: Continente da localização geográfica
* localizacao: Localização geográfica
* data: Data das observações
* total_casos: Total de casos confirmados de COVID-19. As contagens podem incluir casos prováveis, quando reportados.
* novos_casos: Novos casos confirmados de COVID-19. As contagens podem incluir casos prováveis, quando reportados. Em casos raros em que a fonte informa uma variação diária negativa devido a uma correção de dados, essa métrica é definida como NA.
* total_obitos: Total de óbitos atribuídos à COVID-19. As contagens podem incluir óbitos prováveis, quando reportados.
* novos_obitos: Novos óbitos atribuídos à COVID-19. As contagens podem incluir óbitos prováveis, quando reportados. Em casos raros em que a fonte informa uma variação diária negativa devido a uma correção de dados, essa métrica é definida como NA.
* total_casos_por_milhao: Total de casos confirmados de COVID-19 por 1.000.000 de habitantes. As contagens podem incluir casos prováveis, quando reportados.
* novos_casos_por_milhao: Novos casos confirmados de COVID-19 por 1.000.000 de habitantes. As contagens podem incluir casos prováveis, quando reportados.
* total_obitos_por_milhao: Total de óbitos atribuídos à COVID-19 por 1.000.000 de habitantes. As contagens podem incluir óbitos prováveis, quando reportados.
* novos_obitos_por_milhao: Novos óbitos atribuídos à COVID-19 por 1.000.000 de habitantes. As contagens podem incluir óbitos prováveis, quando reportados.
* total_testes: Total de testes para COVID-19
* novos_testes: Novos testes para COVID-19 (calculados apenas para dias consecutivos)
* total_testes_por_mil: Total de testes para COVID-19 por 1.000 habitantes
* novos_testes_por_mil: Novos testes para COVID-19 por 1.000 habitantes
* novos_testes_suavizados: Novos testes para COVID-19 (média móvel de 7 dias). Para países que não reportam dados de testagem diariamente, assume-se que a testagem variou de forma uniforme ao longo dos dias nos períodos em que não houve reporte. Isso gera uma série completa de valores diários, que é então suavizada por meio de uma média móvel de 7 dias.
* novos_testes_suavizados_por_mil: Novos testes para COVID-19 por 1.000 habitantes (média móvel de 7 dias)
* unidade_testes: Unidades utilizadas pela localidade para reportar seus dados de testagem. Um arquivo de país não pode conter unidades mistas. Todas as métricas relacionadas aos dados de testagem utilizam a unidade de teste especificada. As unidades válidas são:
‘people tested’ (número de pessoas testadas),
‘tests performed’ (número de testes realizados; uma única pessoa pode ser testada mais de uma vez no mesmo dia) e
‘samples tested’ (número de amostras testadas; em alguns casos, mais de uma amostra pode ser necessária para realizar um determinado teste).
* indice_rigor: medida composta baseada em 9 indicadores de resposta, incluindo fechamento de escolas, fechamento de locais de trabalho e restrições de viagem, reescalonada para um valor de 0 a 100 (100 = resposta mais rigorosa).
* populacao: População (últimos valores disponíveis). [Consulte aqui](https://github.com/owid/covid-19-data/blob/master/scripts/input/un/population_latest.csv) para a lista completa de fontes.
* densidade_populacional: Número de pessoas dividido pela área terrestre, medido em quilômetros quadrados, ano mais recente disponível.
* idade_mediana: Idade mediana da população, projeção da ONU para 2020.
* populacao_65_ou_mais: Proporção da população com 65 anos ou mais, ano mais recente disponível.
* populacao_70_ou_mais: Proporção da população com 70 anos ou mais em 2015.
* pib_per_capita: Produto Interno Bruto (PIB) a paridade do poder de compra (dólares internacionais constantes de 2011), ano mais recente disponível.
* pobreza_extrema: Proporção da população vivendo em pobreza extrema, ano mais recente disponível desde 2010.
* taxa_mortalidade_cardiovascular: Taxa de mortalidade por doenças cardiovasculares em 2017 (número anual de óbitos por 100.000 pessoas).
* prevalencia_diabetes: Prevalência de diabetes (% da população entre 20 e 79 anos) em 2017.
* fumantes_mulheres: Proporção de mulheres que fumam, ano mais recente disponível.
* fumantes_homens: Proporção de homens que fumam, ano mais recente disponível.
* acesso_lavagem_maos: Proporção da população com instalações básicas para lavagem das mãos no local, ano mais recente disponível.
* leitos_hospitalares_por_mil: Leitos hospitalares por 1.000 habitantes, ano mais recente disponível desde 2010.

* Plotar um gráfico de barras para os 5 países com mais mortes

```python
# Agrupar por localizacao, somar os óbitos
obitos_por_localizacao = df.groupby('localizacao')['total_obitos'].sum().reset_index().copy()

# Ordenar do maior para o menor
obitos_por_localizacao = obitos_por_localizacao.sort_values(by='total_obitos', ascending=False)

# Mostrar o resultado
obitos_por_localizacao = obitos_por_localizacao[obitos_por_localizacao['localizacao'] != 'World']
obitos_por_localizacao = obitos_por_localizacao.head(5)
obitos_por_localizacao
```

```python
import matplotlib.pyplot as plt

plt.Figure(figsize=(10,8))
bars = plt.barh(obitos_por_localizacao['localizacao'],obitos_por_localizacao['total_obitos'])
plt.xlabel('Total de Óbitos')
plt.ylabel('País')
plt.title('Total de Óbitos por País')
for bar in bars:
    width = bar.get_width()
    plt.text(width + 1000, bar.get_y() + bar.get_height()/2, f'{int(width):,}', va='center')
plt.xticks([])
plt.gca().invert_yaxis()  # Inverte para que o maior valor fique no topo
plt.tight_layout()
plt.show();
```

```python
from ipywidgets import interact, fixed, interact_manual
import ipywidgets as widgets
from matplotlib.ticker import FuncFormatter

def plot_scatter(continente):
    plt.clf()

    
    df_filtrado = df_data_maxima[df_data_maxima['continente'] == continente]
    
    # Scatter plot: total de casos x total de óbitos
    plt.scatter(df_filtrado['pib_per_capita'], df_filtrado['total_obitos'], color='blue', alpha=0.7)

    # Melhorar visualização
    plt.xlabel('PIB per Capita')
    plt.ylabel('Total de Óbitos')
    plt.title(f'Total de Óbitos vs PIB per Capita - {continente} {data_maxima.strftime("%d-%m-%Y")}')
    plt.grid(True, linestyle='--', alpha=0.5)

    # Formatar eixo y e x com separador de milhares
    
    formatter = FuncFormatter(lambda x, pos: f'{int(x):,}'.replace(',', '.'))
    plt.gca().xaxis.set_major_formatter(formatter)
    plt.gca().yaxis.set_major_formatter(formatter)

    plt.tight_layout()
    plt.show();

interact(plot_scatter, continente = df_data_maxima['continente'].unique());
```
